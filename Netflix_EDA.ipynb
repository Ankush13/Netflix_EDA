# Data Preprocessing and Cleaning

Lets us first import all the python required libraries to work on this dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Now lets read the csv file using pandas's read_csv function which will be stored as a datafame.

```python
N_df = pd.read_csv('netflix.csv')
```

```python
# To see the first 5 rows of the dataset
N_df.head()
```

```python
# To see the last 5 rows of the dataset
N_df.tail()
```

```python
# Shape property shows the dimension of the dataset( rows, columns )
N_df.shape
```

```python
N_df
```

```python
# to see all the column names
N_df.columns
```

```python
##Lets call the info() function to get the basic information of the data frame like total number of columns with nam ,data types etc.

N_df.info()
```

```python
# to see the aggregate data on dataset
N_df.describe()
```

```python
dup = N_df.duplicated(['title','country', 'type', 'release_year'])
N_df[dup]
```

```python
N_df = N_df.drop_duplicates(['title','country', 'type', 'release_year'])
```

```python
N_df = N_df.drop('show_id', axis = 1)
```

```python
N_df
```

```python
N_df.cast
```

```python
N_df.cast.isna().sum()
```

```python
N_df['cast'] = N_df['cast'].replace(np.nan,'Unknown')
def cast_counter(cast):
    if cast=='Unknown':
        return 0
    else:
        lst=cast.split(', ')
        length=len(lst)
        return length
N_df['number_of_cast'] = N_df['cast'].apply(cast_counter)
N_df['cast'] = N_df['cast'].replace('Unknown',np.nan)
```

```python
N_df
```

```python
N_df = N_df.reset_index()
```

```python
N_df
```

```python
N_df['rating'] = N_df['rating'].fillna(N_df['rating'].mode()[0])
```

```python
N_df.rating.isna().sum()
```

```python
# Lets now check the unique values using nunique function of the dataframe. It will ignore the nan values.
N_df.nunique()
```

```python
# To see the nan values present in particular column we can use isna() or isnull() and apply the aggregate function
# sum() to count the total.


N_df.isna().sum()
# OR
N_df.isnull().sum()
```

```python
# As we observe the above list we can see there are nulls values in director, cast, country, date_added and rating columns.
# Lets remove those and make a copy of the original dataset inorder to not to reflect the changes made in the original one.

NF_df = N_df.copy()
NF_df
```

```python
# Lets remove the director, date and cast columns using the drop function.

NF_df.drop(['director', 'cast', 'date_added'], axis = 1, inplace = True)
```

```python
# Lets check the new dataframe which will have 10 columns

NF_df
```

```python
# But there are again two column left which has nan values : Country and rating
# which we will do using replace function

NF_df['country'].replace(np.NaN, 'United States', inplace = True)
NF_df

# And Same goes for rating column too.
NF_df['rating'].replace(np.NaN, 'TV-MA', inplace = True)
NF_df
```

```python
# Lets verify

NF_df.nunique()

NF_df.isna().sum()
```

# Exploratory Analysis and Visualization

```python
# Lets start analysis the dataset asking certain questions
```

### Q 1. How many TV shows and Movies are present?

```python
N_df
N_df.type
N_df.type.value_counts()
```

```python
# Lets see this in sns with countplot

plt.figure(figsize=(10,8))
sns.countplot(x='rating',hue='type',data=N_df)
plt.title('Comparing frequency between type and rating')
plt.show()
```

```python
# Lets use seaborn and histogram to project the counts

sns.set_style("darkgrid")

plt.hist(N_df.type)
plt.title("Types of Shows and count");
```

```python
# with countplot
sns.countplot(x = 'type', data=N_df);
```

### Q 2. In which country the count is the most?

```python


N_df['country'].value_counts().sort_values(ascending=False)
```

```python
N_df.country
```

```python
# Lets showcase this in graph. we will take limited country to show.


top_ctry = N_df[( N_df['country'] == 'United States' ) | ( N_df['country'] == 'India' ) | ( N_df['country'] == 'United Kingdom' ) | ( N_df['country'] == 'Japan' ) |
                             ( N_df['country'] == 'Canada' ) | ( N_df['country'] == 'Spain' )]
```

```python
top_ctry
```

```python
# GRAPH

plt.figure(figsize=(12,6))
sns.countplot( x = 'country', hue = 'type', data = top_ctry )
plt.title("Comparing between the types that the top countries produce");
```
